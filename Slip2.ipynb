```python
a) Write a Python program to plot 2D graph of the functions f (x) = log10 (x) in the interval [0, 10].
```

```python
from matplotlib.pylab import*
from numpy import *
x=linspace(0,10,100)
f=log10(x)
plot(x,f)
show()
```

```python
(b) Using python, generate 3D surface Plot for the function f (x) = sin x2 + y 2 in the interval [0, 10].
```

```python
from mpl_toolkits import mplot3d
from numpy import *
from pylab import*
x=linspace(0,10,100)
y=linspace(0,10,100)
X,Y=meshgrid(x,y)
f=sin(X**2+Y**2)
ax=axes(projection='3d')
ax.contour3D(X,Y,f,50)
ax.plot(x,y)
show()
```
